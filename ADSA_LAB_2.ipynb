```python
# Convert the Patient List to a Circular Linked list for a Round Robin Check Up System. Implement Insertion and Deletion.

class Node:
    def __init__(self, patient_id, name, age):
        self.patient_id = patient_id
        self.name = name
        self.age = age
        self.next = None


class CircularLinkedList:
    def __init__(self):
        self.last = None


    def insert(self, patient_id, name, age):
        """Insert new patient at the end of the circular list."""
        new_node = Node(patient_id, name, age)
        if self.last is None:
            self.last = new_node
            self.last.next = self.last
        else:
            new_node.next = self.last.next
            self.last.next = new_node
            self.last = new_node
        print(f" Inserted: ID {patient_id} | Name: {name} | Age: {age}")


    def delete(self, patient_id):
        """Delete a patient record by ID."""
        if self.last is None:
            print(" No patients to delete (list is empty).")
            return

        current = self.last.next
        prev = self.last

        while True:
            if current.patient_id == patient_id:
                # Case 1: Only one node
                if current == self.last and current.next == self.last:
                    self.last = None
                # Case 2: Deleting the last node
                elif current == self.last:
                    prev.next = current.next
                    self.last = prev
                # Case 3: Deleting first or middle node
                else:
                    prev.next = current.next
                print(f" Deleted Patient ID {patient_id}")
                return
            prev = current
            current = current.next
            if current == self.last.next:
                break

        print(f" Patient with ID {patient_id} not found.")


    def display(self):
        """Display all patient details."""
        if self.last is None:
            print(" No patients in the list.")
            return

        current = self.last.next
        print("\nCurrent Patient List:")
        print("-" * 40)
        while True:
            print(f"ID: {current.patient_id} | Name: {current.name} | Age: {current.age}")
            current = current.next
            if current == self.last.next:
                break
        print("-" * 40)

    def round_robin_checkup(self, total_rounds):
        """Perform round robin checkup scheduling."""
        if self.last is None:
            print(" No patients for checkup.")
            return

        current = self.last.next
        print("\n Starting Round Robin Checkup")
        print("=" * 40)
        for round_num in range(1, total_rounds + 1):
            print(f"Round {round_num}:")
            print(f"ü©∫ Patient ID: {current.patient_id} | Name: {current.name} | Age: {current.age}")
            print("-" * 40)
            current = current.next
        print(" All scheduled checkup rounds completed!")
        print("=" * 40)


cll = CircularLinkedList()

# Adding patient records
cll.insert(101, "John", 25)
cll.insert(102, "Mary", 30)
cll.insert(103, "David", 28)
cll.insert(104, "Sophia", 22)

# Display all patients
cll.display()

# Delete one patient
cll.delete(103)   # Delete David

# Display after deletion
cll.display()

# Perform 6 checkup rounds
cll.round_robin_checkup(5)
```
